## Считаем P&L по BS

```python
def PnL(interp_df, isCall, K, name_price, name_hedge, last_stock_price=last_stock_price):
    option_price = interp_df[name_price][0]
    T = 61/365
    
    loss=0
    if isCall == True and K < last_stock_price:
        print('Call оказался не выгоден (K < S_T)')
        loss = last_stock_price - K
        
    elif isCall == False and K > last_stock_price:
        print('Put оказался не выгоден (K > S_T)')
        loss = K - last_stock_price
    else:
        print('опцион оказался выгоден')
    
    delta_vpered = interp_df[name_hedge].shift(1)
    delta_vpered[0] = 0
    delta_hedging = np.sum((interp_df[name_hedge] - delta_vpered)*interp_df['S0']*np.exp(-interp_df['r']*(T-interp_df['T']/365)))
    
    r = interp_df['r'][-1]
    delta_last = np.round(interp_df[name_hedge][-1])-interp_df[name_hedge][-1]
    delta_hedging_last = delta_last*last_stock_price*np.exp(-r*T)
    
   
    pnl_hedge = - (delta_hedging + delta_hedging_last)
    
    pnl = (option_price - loss*np.exp(-r*T) + pnl_hedge)
    
    return np.round(pnl,2), np.round(option_price,2), np.round(pnl_hedge,2)
    
```
